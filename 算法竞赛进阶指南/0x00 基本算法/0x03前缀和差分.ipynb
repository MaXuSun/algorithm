1. 前缀和    
* 1维知识点：
  * 对于一个给定的数列A,它的前缀和数列S为: $S[i] = \sum_{j=1}^i A[j]$
  * 一部分和，即数列A某个下标区间内的数的和，可表示为前缀和相减：$sum(l,r)=\sum_{i=l}^r A[i]=S[r] - S[l-1]$
* 2维知识点：
  * $sum[i,j] = sum[i-1,j]+sum[i,j-1]-sum[i-1,j-1]+A[i,j]$
  * 对于一个边长为R的正方形，其值为: $\sum_{x=i-R+1}^i \sum_{y=j-R+1}^j A[x,y] = S[i,j]-S[i-R,j] - S[i,j-R] + S[i-R,j-R]$

2. 差分    
* 定义: $B[1] = A[1], B[i] = A[i] - A[i-1] (2 \leqslant i \leqslant n)$
* 差分和前缀和是一对互逆运算，B的前缀和是A，S的差分是A
* 性质：把序列A的区间[l,r]加d，其差分序列B的变化为：$B_l+d，B_{r+1}-d$，其他位置不变

2.1 IncDec Sequence  
* 问题描述：     
给定一个长度为n $n \leqslant 10^5$ 的数列 ${a_1,a_2,...,a_n}$ ,每次可以选择一个区间[l,r],使下标在这个区间内的数都加1或者都减1。求至少需要多少次操作才能是数列中的数都一样，并求出在保证最小次数的前提下，最终得到的数列可能有多少种。

* 题解：    
  * 求a的差分序列b，其中$b_{n+1}=0$
  * 对于序列a的操作，相当于每次可以选出$b_1,b_2,...,b_{n+1}$中的任意两个数，一个加1，另一个减1。目标是把$b_2,b_3,...,b_n$变成全零。最终得到的数列a就是由n个$b-1$构成的
  * 从$b_1,b_2,...,b_{n+1}$中任选两个数的方法可分为四类：
    * 选$b_i$和$b_j$，其中 $2 \leqslant i,j \leqslant n$ 。应保证在$b_i$和$b_j$一正一负的前提下，尽量多地采用这种操作，更快接近目标
    * 选$b_1$和$b_j$，其中 $2 \leqslant j \leqslant n$ 。
    * 选$b_i$和$b_{n+1}$，其中 $2 \leqslant i \leqslant n$ 。
    * 选$b_1$和$b_{n+1}$。这种选择没有意义，不会改变$b_2,b_3,...,b_n$的值，相当于浪费了一次操作，一定不是最优解
* 答案：
  * 设$b_2,b_3,...,b_n$中正数总和为p,负数总和的绝对值为q。首先尽量执行第一类操作，可执行 $min(p,q)$次，剩下|p-q|个未配对，可执行2，3选择策略，共|p-q|次
  * 综上，最少需要$min(p,q)+|p-q|=max(p,q)$次，根据|p-q|次第2，3类操作的选择情况，能产生|p-q|+1中不同的$b_1$的值，即最后得到的序列a可能有|p-q|+1种。

2.1 Tallest Cow    
* 问题描述：     
有N头牛站成一行。如果两头牛中间牛的身高都比她们矮，它们可以互相看见。现在，我们只知道其中最高的牛是第 头，它的身高是 H, 不知道剩余 N-1牛的身高。但是，我们还知道 对关系，每对关系都明了某两头牛$A_i,B_i$可以相互看见。求每头牛的身高最大可能是多少。
* 解题：     
题中M对关系实际是牛之间身高的相对大小，建立一个数组C，初始化全0，若一条关系指明$A_i$和$B_i$可以相互看见，则把C中下标$A_i+1$到$B_i-1$的数都减去1。因为P牛是最高的，所以最终C[P]一定为0。其他牛与第P头牛的身高差距几句体现在数组C中。换言之，最后第i投牛的身高就等于H+C[i]
* 优化
如果把C中下标$A_i+1$到$B_i-1$都执行减1操作，则复杂度较高。可以新建一个数组D,领D[$A_i+1$]减1，D[$B_i$]加上1。这时$C[i] = \sum_{j=1}^i D[j]$
