```python
def solve(digits,res,stack,visit):
    if len(stack) == 3:
        if stack[-1] % 2 == 0:
            temp = stack[0]*100+stack[1]*10+stack[-1]
            if len(res)==0 or temp > res[-1]:
                res.append(temp)
        return
    
    for i in range(len(digits)):
        if visit[i]:
            continue
        if digits[i] == 0 and len(stack) == 0:
            continue

        visit[i] = True
        stack.append(digits[i])
        solve(digits,res,stack,visit)
        stack.pop()
        visit[i] = False

def findEvenNumbers(digits):
    res = []
    stack = []
    visit = [False] * len(digits)
    digits.sort()
    solve(digits,res,stack,visit)
    return res

print(findEvenNumbers([0,2,0,0]))
```

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def deleteMiddle(self, head):
        Head = ListNode(next = head)
        p = Head
        q = Head
        flag = 1
        while q is not None and q.next is not None:
            p = p.next
            q = q.next.next
        if q is None:
            flag = 0
        if flag == 0:
            p.next = p.next.next
        else:
            q = Head
            while q.next != p:
                q = q.next
            q.next = p.next
        return Head.next
        
```
