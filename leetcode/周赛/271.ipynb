```python
def countPoints(rings):
    res = 0
    store = [[0]*3 for _ in range(10)]
    for i in range(len(rings)//2):
        s = rings[2*i]
        if s == 'R':
            s = 0
        elif s == 'G':
            s = 1
        else:
            s = 2
        num = int(rings[2*i+1])
        store[num][s] = 1
    for i in range(9):
        if sum(store[i]) == 3:
            res += 1
    return res

print(countPoints("B0B6G0R6R0R6G9"))
```

```python
def subArrayRanges(nums):
    length = len(nums)
    store = [[0]*length for _ in range(length)]
    max_val = [[0]*length for _ in range(length)]
    min_val = [[0]*length for _ in range(length)]
    
    res = 0
    # 初始化横轴和纵轴
    for j in range(length):
        store[0][j] = nums[j]-nums[0]
        max_val[j][j] = nums[j]
        min_val[j][j] = nums[j]
        if j-1 >= 0:
            store[j-1][j] = nums[j]-nums[j-1]
            max_val[0][j] = max(max_val[0][j-1],nums[j])
            min_val[0][j] = min(min_val[0][j-1],nums[j])

            max_val[j-1][j] = max(nums[j-1],nums[j])
            min_val[j-1][j] = min(nums[j-1],nums[j])
    # print(max_val[0])
    for i in range(length):
        for j in range(length):
            if i >= j: continue
            if nums[j] > max_val[i][j-1]:
                store[i][j] = nums[j] - min_val[i][j-1]
                max_val[i][j] = nums[j]
                min_val[i][j] = min_val[i][j-1]
            elif nums[j] < min_val[i][j-1]:
                store[i][j] = max_val[i][j-1] - nums[j]
                min_val[i][j] = nums[j]
                max_val[i][j] = max_val[i][j-1]
            else:
                store[i][j] = store[i][j-1]
                min_val[i][j] = min_val[i][j-1]
                max_val[i][j] = max_val[i][j-1]
            res += store[i][j]

    for i in range(length):
        for j in range(length):
            print(store[i][j],end=' ')
        print()
    print()
    for i in range(length):
        for j in range(length):
            print(max_val[i][j],end=' ')
        print()
    print()
    for i in range(length):
        for j in range(length):
            print(min_val[i][j],end=' ')
        print()
    return res
    
print(subArrayRanges([22,-12,88,56,-3,-96]))
    
```

```python
def minimumRefill(plants,capacityA,capacityB):
    storeA = capacityA
    storeB = capacityB
    i,j = 0,len(plants)-1
    res = 0
    print(plants[28])
    while i < j:
        while i<j and capacityA >= plants[i]:
            capacityA-=plants[i]
            i+=1
        while i < j and capacityB >= plants[j]:
            capacityB-=plants[j]
            j-=1
        if i < j:       # 他俩依然没有相遇
            if capacityA < plants[i] and capacityA != storeA:
                res +=1
                capacityA = storeA
            if capacityB < plants[j] and capacityB != storeB:
                res+=1
                capacityB = storeB
        print(res,i,j,capacityA,capacityB)
    if capacityA < plants[i] and capacityB < plants[j]:     # 如果他俩都不够浇水，肯定有一个要补充水
        res+=1

    return res

print(minimumRefill([923,65,50,788,446,122,987,355,528,590,69,107,656,676,454,604,213,299,969,239,341,940,690,195,884,135,499,38,440,857,404,266,786,915,564,246,271,977,775,978,166,636,494,284,435,630,584,590,401,167],1455,1343))
```
